```python
import matplotlib.pyplot as plt                       #importing pyplot module from matplotlib library
months=["jan","feb","march","april","may","june","july","aug","sept","oct","nov","dec"]
London=[6,6,8,11,14,17,19,18,16,13,9,6]
NewDelhi=[14,17,23,29,34,34,31,30,29,26,20,16]
plt.plot(months,London,color="r",label="London")    #plotting the first line chart in the figure for London with red line
plt.plot(months,NewDelhi,color="g",label="New Delhi")#plotting the second line chart in the same figure for New Delhi with blue line
plt.title("Temperature (London v/s New Delhi)")      #giving a title to the plot
plt.xlabel("Months",fontsize=12)                    #giving label to x axis of the plot
plt.ylabel("Avg Temperature in Celcius",fontsize=12)#giving label to y axis of the plot
plt.ylim(0,35)                                      #specifying the start and end of tick marks along the y axis
plt.legend()                                        #displaying the legend
plt.show()                                          #displaying the figure
```

```python
import matplotlib.pyplot as plt                                    #importing pyplot module from matplotlib library
import numpy as np                                                 #importing numpy library
y1=[120,130,100,80,90]                                             #list of frequency for first group (which is 2018 here)
y2=[90,120,80,120,100]                                             #list of frequency for the second group (2019)
barwidth=0.35                                                      #assigning the width for each bar
bar1=np.arange(len(y1))                                            #position of each bar of the first group
bar2=[x+barwidth for x in bar1]                                    #position of each bar of the second group
plt.bar(bar1,y1,color="r",width=barwidth,label=2018)               #plotting bar chart for the first group (i.e, 2018)
plt.bar(bar2,y2,color="g",width=barwidth,label=2019)               #plotting bar chart for the second group (i.e, 2019)
plt.xticks([r for r in range(len(y1))], ['A', 'B', 'C', 'D', 'E']) #tick marks for each grouped bar 
plt.title("2018 vs 2019")
plt.xlabel("Students")
plt.ylabel("Mark")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np                                                 #importing numpy library as np
y1=[80,65,100,80,90]
y2=[90,52,80,85,100]
barwidth=0.35
bar1=np.arange(len(x1))
plt.bar(bar1,y1,color="r",width=barwidth,label="Semester1")         #plotting bar chart for the first group (i.e,Semester1)  
plt.bar(bar1,y2,color="g",bottom=y1,width=barwidth,label="Semester2")#plotting bar chart for the second group over first
plt.xticks([r for r in range(len(y1))], ['Maths', 'Biology', 'Chemistry', 'Physics', 'English']) 
plt.yticks(np.arange(0,250,20))
plt.title("Marks in the year 2019")
plt.xlabel("Subject")
plt.ylabel("Mark")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
months=["jan","feb","march","april","may","june","july","aug","sept","oct","nov","dec"]
London=[6,6,8,11,14,17,19,18,16,13,9,6]
NewDelhi=[14,17,23,29,34,34,31,30,29,26,20,16]
fig,ax=plt.subplots( )
ax.plot(months,London,color="r",label="London")    #plotting the first line chart in the figure for London with red line
ax.plot(months,NewDelhi,color="g",label="New Delhi")#plotting the second line chart in the same figure for New Delhi with blue line
ax.set_title("Temperature (London v/s New Delhi)")      #giving a title to the plot
ax.set_xlabel("Months",fontsize=12)                    #giving label to x axis of the plot
ax.set_ylabel("Avg Temperature in Celcius",fontsize=12)#giving label to y axis of the plot
ax.set_ylim(0,35)                                      #specifying the start and end of tick marks along the y axis
ax.legend()                                        #displaying the legend
plt.show()                                          #displaying the figure
```
